### eliminate outliers / surprise scores

```python
model_input[0][-1]
```

```python
model_input.shape, model_output.shape
```

```python
model_input_no_outlier = model_input.copy()
model_output_no_outlier = model_output.copy()

index_to_remove = list()

for i, value in enumerate(zip(model_input, model_output)):
    model, score = value
    
    print(model[-1], score)
    if model[-1][0] > model[-1][1] and score[0] < score[1]:
        print('outlier1')
        index_to_remove.append(i)
        #model_input_no_outlier = np.delete(model_input_no_outlier, i, axis=0)
        #model_output_no_outlier = np.delete(model_output_no_outlier, i, axis=0)
        #print(f'removed outlier index {i}')
    
    elif model[-1][1] > model[-1][0] and score[1] < score[0]:
        print('outlier2')
        index_to_remove.append(i)
        #model_input_no_outlier = np.delete(model_input_no_outlier, i, axis=0)
        #model_output_no_outlier = np.delete(model_output_no_outlier, i, axis=0)
        #print(f'removed outlier index {i}')

model_input_no_outlier = np.delete(model_input, index_to_remove, axis=0)
model_output_no_outlier = np.delete(model_output, index_to_remove, axis=0)
```

```python
model_input_no_outlier.shape, model_output_no_outlier.shape
```

```python
model_input_ratio = list()

for model in model_input_no_outlier:
    #print(model.shape)
    _per_model = list()
    for i in range(model.shape[0]):
        #print(model[i])
        _t = (model[i][0] + _epsilon) / (model[i][1] + _epsilon)
        _per_model.append(_t)

    _per_model = np.array(_per_model)
    model_input_ratio.append(_per_model)

model_input_ratio = np.array(model_input_ratio)
```

## gradient boosting

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.ensemble import GradientBoostingRegressor
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.multioutput import MultiOutputRegressor

gb_regressor = GradientBoostingRegressor(random_state=42)

# Initialize the MultiOutputRegressor
multioutput_regressor = MultiOutputRegressor(gb_regressor)

multioutput_regressor.fit(X_train, y_train)
```

```python
y_train.shape, X_train.shape
```

```python
# Predict on the test set
y_pred_gbr = multioutput_regressor.predict(X_test)
```

```python
y_test_series = pd.DataFrame(y_test, columns=['odd1','odd2'])
y_pred_gbr_series = pd.DataFrame(y_pred_gbr, columns=['pred1','pred2'])
res_gbr = pd.concat([y_test_series, y_pred_gbr_series], axis=1)
```

```python
res_gbr
```

```python
mean_squared_error(y_test, y_pred_gbr)
```
